```python
# importar labelEncoder desde sklearn
from sklearn.preprocessing import LabelEncoder
```

```python
#Crear un objeto label encoder
gle = LabelEncoder()

#Transformación a lables la columna Genre utilizando el objeto gle
genre_labels = gle.fit_transform(df['Genre'])
```

```python
genre_labels
```

```python
# Añadir la nueva columna al dataset
# df['Genre_Labels'] = genre_labels
```

```python
# Crear un diccionario con el mapeo de indices y valores

genre_mappings = {k:v for k,v in enumerate(gle.classes_)}
genre_mappings
```

```python
df['Genre_Labels'] = genre_labels
```

```python
df.loc[:,['Name','Platform','Genre','Genre_Labels']].head(10)
```

#Dummy Coding

```python
# Crear las columnas dummies para la columna Genre (Está dentro de pandas)

genre_dummies = pd.get_dummies(df['Genre'])
genre_dummies
```

```python
# Crear un nuevo dataset (Copia) que contenga las columnas más importantes + las dummies

df_final = pd.concat([df[['Name','Publisher','Year','Genre']],genre_dummies], axis = 1)
df_final.head(10)
```

#Feature Hashing

```python
print(np.unique(df['Platform']))
print(len(np.unique(df['Platform'])))
```

Aplicaremos Feature Hashing para codificar la columna Platform, debido a que contiene demasiadas categorías para aplicar get_dummies

```python
# Importar la libreria de Feature Hshing
from sklearn.feature_extraction import FeatureHasher

#Crear un objeto FeatureHasher
fh = FeatureHasher(n_features = 10, input_type = 'string')

# Realizamos la codificación
hashed_features = fh.fit_transform(df['Platform'])

#Convertir en array el resultado
hashed_features = hashed_features.toarray()

#Crear un nuevo dataframe con la columna codificada
df_final2 = pd.concat([df[['Name','Publisher', 'Year', 'Genre','Platform']], pd.DataFrame(hashed_features)], axis=1)
df_final2.head(10)
```
